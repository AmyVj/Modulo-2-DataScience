```python
my_conn = create_engine('mysql+pymysql://root:root1234@localhost:3306/henry_2')

if my_conn:
    print('CONEXION EXITOSA!!!!')
```

```python
# NAME -> Nombre de la tabla que queremos cargar
# if_exists ['replace','append','fail'] -> Si la tabla ya existe, que debe pasar con la carga? 
# if_exists = 'replace' -> Reemplaza los valores viejos por los nuevos que estoy por cargar
# if_exists = 'append' -> Suma los nuevos valores a la tabla que ya existe
# if_exists = 'fail' -> Tira error si la tabla ya existe
# index -> 'False' si queremos que no se cree una columna sin nombre con los valores de los indices

# con -> Conexion entre Python y la Base de Datos creada a traves de SQLAlchemy

# PATH = ''motorsql+motorpython://USER:PASSWORD@HOST:PORT/DATABASE'
# PATH -> 'mysql+pymysql://root:root1234@localhost:3306/henry'
# create_engine('mysql+pymysql://root:root1234@localhost:3306/henry')


df.to_sql(name='alumnos', con=my_conn, if_exists='replace', index=False)
```

```python
# Para leer una query desde pandas utilizando SQLAlchemy

query = 'SELECT * FROM persona;'
df2 = pd.read_sql_query(sql=query, con=my_conn)
df2.to_csv('personas.csv')
```
